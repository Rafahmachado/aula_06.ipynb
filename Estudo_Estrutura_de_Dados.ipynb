# Tuplas #
## Estrutura similar a lista,com apenas uma diferença:os elementos inseridos em uma tupla não podem ser alterados,diferente de uma lista onde podem ser alterados livremente ##


```python
#Tupla é utilizada para armazenar dados de tipos diferentes
#Podem ser utilizadas como chave para um dicionário,já que seus elementos são imutáveis.Com a lista não é possível.
nome_da_tupla = (1, 2, 3) #tupla de inteiros
nome_da_tupla = (1, "olá", 1.5) #tupla heterogênea
```

# Sets #

## No Python,os sets são uma coleção de itens desordenada,parcialmnente imutavel e que não podem conter elementos duplicados ##

```python
# Sets são utilizados com operações mataemáticas de união ,interseção e diferença simétrica
# Declarando Sets,utiliza-se chaves{}para determinar os elementos do set
nome_do_set = {1, 2, 3, 4}
```

```python
# Outra forma seria o método set presente no set
nome_do_set = set([1, 2, 3, 4])
```

# Dicionários #

## Os dicionários são coleções de itens desordenados com uma diferença:um elemento dentro de um dicionário possui uma chave atrelada a ele ou sej aum identificador ##

```python
# Declarando Dicionários:
# Utilizando chaves {} e separando os elementos das chaves com dois pontos,
# e os outros elementos por vírgula.

nome_do_dicionario = {1: "João", 2: "José"}
nome_do_dicionario = {"nome": "João", "sobrenome": "Silva"}
```

```python
# Podemos declarar também utilizando o método dict() com o dicionário sendo passado como parâmentro:
nome_do_dicionario = dict({1: "João", 2: "José"})
nome_do_dicionario = dict({"nome": "João", "sobrenome": "Silva"})
```
